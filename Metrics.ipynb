```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
met_df_51_4 = pd.read_csv("/Users/joel/Projects/optidb/upgrade/metrics51-4.csv")
met_df_23_10 = pd.read_csv("/Users/joel/Projects/optidb/upgrade/metrics23-10.csv")
met_df_41_12 = pd.read_csv("/Users/joel/Projects/optidb/upgrade/metrics41-12.csv")
met_df_30_5 = pd.read_csv("/Users/joel/Projects/optidb/upgrade/metrics30-5.csv")
met_df_29_8 = pd.read_csv("/Users/joel/Projects/optidb/upgrade/metrics29-8.csv")
```

```python
plt.figure(figsize=(12,7.5))

ax = plt.subplot(111)

ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False) 

plt.plot([i for i in range(35)],met_df_51_4['avg'].values.tolist()[:35],label="$N=51, \epsilon=4.0kT$")
plt.plot([i for i in range(35)],met_df_23_10['avg'].values.tolist()[:35],label="$N=23, \epsilon=10.0kT$")
plt.plot([i for i in range(35)],met_df_29_8['avg'].values.tolist()[:35],label="$N=29, \epsilon=8.0kT$")

ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left()   
plt.tick_params(axis="both", which="both", bottom="off", top="off",    
                    labelbottom="on", left="off", right="off", labelleft="on")  

plt.xlabel("Generation",fontsize=14)
plt.ylabel("Mean Population Fitness",fontsize=14)
plt.legend(fontsize=14,loc='lower right')
plt.show()
```
